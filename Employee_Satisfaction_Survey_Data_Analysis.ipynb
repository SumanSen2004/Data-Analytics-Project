# ****Employee Survey Data Analysis using Python (NumPy, Pandas, and Matplotlib)****

## **1. Importing Necessary Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## **2. Loading and Understanding the Dataset**

#### **I. Load Dataset**

```python
df = pd.read_csv("employee_survey.csv")
```

#### **II. Display First Few Rows**

```python
print(df.head())
```

#### **III. Check for Missing Values**

```python
print("Missing Values:\n", df.isnull().sum())
```

#### **IV. Fill Missing Values with 0**

```python
df.fillna(0, inplace=True)
```

## **3. Employee Satisfaction Analysis**

****Question: What is the distribution of employee satisfaction levels?****

**Description: Using NumPy and Matplotlib to analyze the distribution.**

```python
# Using NumPy to get statistical insights
mean_satisfaction = np.mean(df['satisfaction_level'])
median_satisfaction = np.median(df['satisfaction_level'])
std_satisfaction = np.std(df['satisfaction_level'])

print(f"Mean Satisfaction Level: {mean_satisfaction:.2f}")
print(f"Median Satisfaction Level: {median_satisfaction:.2f}")
print(f"Standard Deviation: {std_satisfaction:.2f}")

# Histogram
plt.figure(figsize=(6,3))
plt.hist(df['satisfaction_level'], bins=5, color='blue', alpha=0.7, edgecolor='black')
plt.axvline(mean_satisfaction, color='red', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(median_satisfaction, color='green', linestyle='dashed', linewidth=2, label='Median')
plt.title('Employee Satisfaction Level Distribution')
plt.xlabel('Satisfaction Level')
plt.ylabel('Number of Employees')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## **4. Department-wise Employee Count**

****Question: How many employees are in each department?****

**Description: Using Pandas and Matplotlib for visualization.**

```python
# Using Pandas for department count
dept_counts = df['dept'].value_counts()
print(dept_counts)

# Bar Chart
plt.figure(figsize=(6,3))
dept_counts.plot(kind='bar', color='green', alpha=0.7, edgecolor='black')
plt.title('Number of Employees in Each Department')
plt.xlabel('Department')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## **5. Salary Level Distribution**

****Question: How many employees have low, medium, and high salaries?****

**Description: Analyzing salary distribution using Pandas and Matplotlib.**

```python
# Using Pandas to count salary distribution
salary_counts = df['salary'].value_counts()
print(salary_counts)

# Pie Chart
plt.figure(figsize=(6,3))
salary_counts.plot(kind='pie', autopct='%1.1f%%', colors=['red', 'orange', 'green'])
plt.title('Distribution of Employee Salaries')
plt.ylabel('')  # Hide y-label
plt.show()
```

## **6. Relationship Between Satisfaction Level and Work Hours**

****Question: Is there any relationship between job satisfaction and average monthly working hours?****

**Description: Using NumPy for correlation and Matplotlib for visualization.**

```python
# Using NumPy to calculate correlation
correlation = np.corrcoef(df['average_montly_hours'], df['satisfaction_level'])[0, 1]
print(f"Correlation between Work Hours and Satisfaction Level: {correlation:.2f}")

# Scatter Plot
plt.figure(figsize=(6,3))
plt.scatter(df['average_montly_hours'], df['satisfaction_level'], color='purple', alpha=0.6)
plt.title('Satisfaction Level vs. Average Monthly Hours')
plt.xlabel('Average Monthly Hours')
plt.ylabel('Satisfaction Level')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

## **7. Employee Promotions Over the Last 5 Years**

****Question: How many employees have received a promotion in the last 5 years?****

**Description: Analyzing promotion data using Pandas and Matplotlib.**

```python
# Using Pandas to count promotions
promotion_counts = df['promotion_last_5years'].value_counts()
print(promotion_counts)

# Bar Chart
plt.figure(figsize=(6,3))
promotion_counts.plot(kind='bar', color=['gray', 'blue'], alpha=0.7, edgecolor='black')
plt.title('Promotion Status in Last 5 Years')
plt.xlabel('Promotion (0 = No, 1 = Yes)')
plt.ylabel('Number of Employees')
plt.xticks(ticks=[0,1], labels=['Not Promoted', 'Promoted'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## **8. Work Accidents Analysis**

****Question: How many employees have experienced a work accident?****

**Description: Analyzing work accidents using Pandas and Matplotlib.**

```python
# Using Pandas to count work accidents
accident_counts = df['Work_accident'].value_counts()
print(accident_counts)

# Bar Chart
plt.figure(figsize=(6,3))
accident_counts.plot(kind='bar', color=['red', 'green'], alpha=0.7, edgecolor='black')
plt.title('Work Accident Occurrences')
plt.xlabel('Work Accident (0 = No, 1 = Yes)')
plt.ylabel('Number of Employees')
plt.xticks(ticks=[0,1], labels=['No Accident', 'Had Accident'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## **9. Time Spent in Company vs. Satisfaction Level**

****Question: Does the number of years spent in the company affect job satisfaction?****

**Description: Analyzing satisfaction trends over time.**

```python
# Using Pandas to get mean satisfaction per year
satisfaction_by_year = df.groupby('time_spend_company')['satisfaction_level'].mean()
print(satisfaction_by_year)

# Line Chart
plt.figure(figsize=(6,3))
satisfaction_by_year.plot(kind='line', marker='o', color='blue', linestyle='-')
plt.title('Average Satisfaction Level by Years in Company')
plt.xlabel('Years Spent in Company')
plt.ylabel('Average Satisfaction Level')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

### **Conclusion:**

****Job Satisfaction:**** Most employees have moderate satisfaction levels.

****Department Count:**** IT, Sales, and Finance have more employees than HR and Marketing.

****Salary Levels:**** Most employees fall into the low or medium salary categories.

****Work Hours vs. Satisfaction:**** The correlation suggests that excessive work hours reduce satisfaction.

****Promotions:**** Very few employees receive promotions over 5 years.

****Work Accidents:**** Work accidents are not highly frequent but still a concern.

****Time Spent in Company:**** Satisfaction tends to decline as employees spend more years in the company.

This analysis effectively highlights key workplace trends using NumPy, Pandas, and Matplotlib.
